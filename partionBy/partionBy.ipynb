```
A = sc.parallelize(range(1000000))
```

```
print(A.getNumPartitions())
```

```
# Repartition

B = A.repartition(10)
print(B.getNumPartitions())
```

```
C = sc.parallelize(range(1000000)) \
  .map(lambda x: (x,x)).partitionBy(10)
print(C.glom().map(len).collect())
```

```
# Suppose we use filter to select some elements of C
D = C.filter(lambda pair: pair[0]%5 == 0)
# get no. of partitions
print(D.glom().map(len).collect())
```

```
# Only two workers of D have values, so need to repartition, 

E = D.map(lambda pair: (pair[1]/10, pair[1])).partitionBy(10)
print(E.glom().map(len).collect())
```
